```python
from langgraph.store.memory import InMemoryStore
from langchain_openai import OpenAIEmbeddings
from langchain_groq import ChatGroq
from CONFIG import GROQ_MODEL, TEMPERATURE
from dotenv import load_dotenv
```

```python
load_dotenv()

embedding_llm = OpenAIEmbeddings(
    model='text-embedding-3-small'
)

llm = ChatGroq(model=GROQ_MODEL, temperature=TEMPERATURE)
```

```python
namespace = ("user", "profile")
```

```python
store = InMemoryStore(index={'embed': embedding_llm, 'dims': 1536})
```

```python
store.put(namespace, '0', {'data': 'my name is adnan saeed'})
store.put(namespace, '1', {'data': 'i like dark mode while doing coding'})
store.put(namespace, '2', {'data': 'i am working as an AI Engineer'})
store.put(namespace, '3', {'data': 'i am workin on ai agents'})
store.put(namespace, '4', {'data': 'i live in pakistan'})
store.put(namespace, '5', {'data': 'i leaned machine learning and deep learning'})
store.put(namespace, '6', {'data': 'i want to move UAE'})
```

```python
items = store.search(namespace, query="what's my plan?", limit=1)
```

```python
for item in items:
    print(item.value['data'])
```
